```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# import and reading dataset csv file

iris_df = pd.read_csv('Iris.csv')
```

```python
# Viewing dataset
iris_df.head()
```

```python
# Drop id column
iris_df.drop('Id', axis=1, inplace=True)
```

```python
iris_df.head()
```

```python
iris_df.shape
```

```python
#finding number of missing values

iris_df.isnull().sum()
```

```python
iris_df.info()
iris_df.describe()
```

```python
scaler = StandardScaler()
iris_df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] = scaler.fit_transform(iris_df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']])
```

```python
X = iris_df.drop('Species', axis= 1)
y = iris_df['Species']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
# Logistic regression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
# Select features
x = 'SepalLengthCm'
y = 'SepalWidthCm'

# Plot for each species
plt.scatter(iris_df[iris_df['Species']=='Iris-setosa'][x], iris_df[iris_df['Species']=='Iris-setosa'][y], color='red', label='Setosa')
plt.scatter(iris_df[iris_df['Species']=='Iris-versicolor'][x], iris_df[iris_df['Species']=='Iris-versicolor'][y], color='green', label='Versicolor')
plt.scatter(iris_df[iris_df['Species']=='Iris-virginica'][x], iris_df[iris_df['Species']=='Iris-virginica'][y], color='blue', label='Virginica')

# Chart labels  
plt.title('Iris Species Classification')
plt.xlabel(x)
plt.ylabel(y)

# Legend
plt.legend()

plt.show()
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
print('Accuracy:', accuracy_score(y_test, y_pred))
```

```python
# Classification Report
from sklearn.metrics import classification_report
print('Classification Report:\n', classification_report(y_test, y_pred))
```
